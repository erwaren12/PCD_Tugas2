```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('Image.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
h, s, v = cv2.split(hsv)
v_float = v.astype(np.float32)

v_log_36 = np.clip(36 * np.log1p(v_float), 0, 255).astype(np.uint8)
v_log_50 = np.clip(50 * np.log1p(v_float), 0, 255).astype(np.uint8)

hasil_36 = cv2.cvtColor(cv2.merge([h, s, v_log_36]), cv2.COLOR_HSV2RGB)
hasil_50 = cv2.cvtColor(cv2.merge([h, s, v_log_50]), cv2.COLOR_HSV2RGB)

plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1); plt.imshow(img_rgb); plt.title('Citra Asli'); plt.axis('off')
plt.subplot(1, 3, 2); plt.imshow(hasil_36); plt.title('Log c=36'); plt.axis('off')
plt.subplot(1, 3, 3); plt.imshow(hasil_50); plt.title('Log c=50'); plt.axis('off')
plt.tight_layout()
plt.show()
```
