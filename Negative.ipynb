```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('Image.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
h, s, v = cv2.split(hsv)

v_neg_255 = 255 - v
v_neg_128 = np.clip(128 - v.astype(np.int16), 0, 255).astype(np.uint8)

hasil_255 = cv2.cvtColor(cv2.merge([h, s, v_neg_255]), cv2.COLOR_HSV2RGB)
hasil_128 = cv2.cvtColor(cv2.merge([h, s, v_neg_128]), cv2.COLOR_HSV2RGB)

plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1); plt.imshow(img_rgb); plt.title('Citra Asli'); plt.axis('off')
plt.subplot(1, 3, 2); plt.imshow(hasil_255); plt.title('Negative L=255'); plt.axis('off')
plt.subplot(1, 3, 3); plt.imshow(hasil_128); plt.title('Negative L=128'); plt.axis('off')
plt.tight_layout()
plt.show()
```
