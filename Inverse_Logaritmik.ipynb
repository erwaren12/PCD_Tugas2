```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('Image.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
h, s, v = cv2.split(hsv)
v_float = v.astype(np.float32)

v_inv_30 = np.clip(30 * np.log1p(255 - v_float), 0, 255).astype(np.uint8)
v_inv_50 = np.clip(50 * np.log1p(255 - v_float), 0, 255).astype(np.uint8)

hasil_30 = cv2.cvtColor(cv2.merge([h, s, v_inv_30]), cv2.COLOR_HSV2RGB)
hasil_50 = cv2.cvtColor(cv2.merge([h, s, v_inv_50]), cv2.COLOR_HSV2RGB)

plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1); plt.imshow(img_rgb); plt.title('Citra Asli'); plt.axis('off')
plt.subplot(1, 3, 2); plt.imshow(hasil_30); plt.title('Inv Log c=30'); plt.axis('off')
plt.subplot(1, 3, 3); plt.imshow(hasil_50); plt.title('Inv Log c=50'); plt.axis('off')
plt.tight_layout()
plt.show()
```
