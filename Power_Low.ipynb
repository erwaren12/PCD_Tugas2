```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('Image.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
h, s, v = cv2.split(hsv)
v_float = v.astype(np.float32)

v_pow_1 = np.clip(1.0 * np.power(v_float, 0.9), 0, 255).astype(np.uint8)
v_pow_2 = np.clip(1.5 * np.power(v_float, 0.7), 0, 255).astype(np.uint8)

hasil_pow_1 = cv2.cvtColor(cv2.merge([h, s, v_pow_1]), cv2.COLOR_HSV2RGB)
hasil_pow_2 = cv2.cvtColor(cv2.merge([h, s, v_pow_2]), cv2.COLOR_HSV2RGB)

plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1); plt.imshow(img_rgb); plt.title('Citra Asli'); plt.axis('off')
plt.subplot(1, 3, 2); plt.imshow(hasil_pow_1); plt.title('Power Law c=1.0, y=0.9'); plt.axis('off')
plt.subplot(1, 3, 3); plt.imshow(hasil_pow_2); plt.title('Power Law c=1.5, y=0.7'); plt.axis('off')
plt.tight_layout()
plt.show()
```
